```python
class StudentDatabase:
    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter student name: ")
        roll_number = input("Enter roll number: ")
        marks = input("Enter marks: ")
        grade = input("Enter grade: ")
        student = {
            'name': name,
            'roll_number': roll_number,
            'marks': marks,
            'grade': grade
        }
        self.students.append(student)
        print("Student added successfully.")

    def remove_student(self):
        roll_number = input("Enter roll number of the student to remove: ")
        for student in self.students:
            if student['roll_number'] == roll_number:
                self.students.remove(student)
                print("Student removed successfully.")
                return
        print("No student found with the given roll number.")

    def search_student(self):
        roll_number = input("Enter roll number of the student to search: ")
        for student in self.students:
            if student['roll_number'] == roll_number:
                print("Student found:")
                print("Name: {student['name']}")
                print("Roll Number: {student['roll_number']}")
                print("Marks: {student['marks']}")
                print("Grade: {student['grade']}")
                return
        print("No student found with the given roll number.")

    def update_student(self):
        roll_number = input("Enter roll number of the student to update: ")
        for student in self.students:
            if student['roll_number'] == roll_number:
                print("Enter new details:")
                name = input("Enter updated name: ")
                roll_number = input("Enter roll number: ")
                marks = input("Enter marks: ")
                grade = input("Enter updated grade: ")
                student['name'] = name
                student['roll_number'] = roll_number
                student['marks'] = marks
                student['grade'] = grade
                print("Student details updated successfully.")
                return
        print("No student found with the given roll number.")

    def display_all_students(self):
        if len(self.students) == 0:
            print("No students found in the database.")
        else:
            print("Student Database:")
            for student in self.students:
                print("Name:", student['name'])
                print("Roll Number:", student['roll_number'])
                print("Marks:", student['marks'])
                print("Grade:", student['grade'])
                print()

    def useprogram(self):
        while True:
            print("\n--- Student Database Management System ---")
            print("1. Add Student")
            print("2. Remove Student")
            print("3. Search Student")
            print("4. Update Student Details")
            print("5. Display All Students")
            print("6. Exit")

            choice = input("Enter your choice (1-6): ")

            if choice == "1":
                self.add_student()

            elif choice == "2":
                self.remove_student()

            elif choice == "3":
                self.search_student()

            elif choice == "4":
                self.update_student()

            elif choice == "5":
                self.display_all_students()

            elif choice == "6":
                print("Exiting the program...")
                break

            else:
                print("Invalid choice. Please enter a valid option (1-6).")

db = StudentDatabase()
db.useprogram()
```
